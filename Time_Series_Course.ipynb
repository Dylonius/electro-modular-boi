```python
data = pd.read_csv("Course+DataSet.csv",index_col="DATE",parse_dates = True)
```

```python
import matplotlib.pyplot as plt
start_date = "01-01-2010"
end_date = "01-01-2017"
timelim = (data.index >= start_date) & (data.index < end_date)
datatimelim = data.loc[timelim]
datatimelim.plot(figsize = (12,8))
xcoords = ["01-01-2011","01-01-2012","01-01-2013","01-01-2014","01-01-2015","01-01-2016"]
for xc in xcoords:
  plt.axvline(x=xc, color="black",linestyle="--")
```
